# Representation of uni-variate numerical data.

# 1). Boxplot / Kernel Density Estimate(KDE).

```python
For Performinhg the computation task on the datasets it is assumed that the data are clustered around some central values.
To create a box and whisker plot, we start by ordering our data (that is, putting the values) in numerical order, if they
arenot ordered already. Then we find the median of our data.The median divides the data into two halves.To divide the data
into quarters, we then find the medians of these two halves.

```

```python
sns.boxplot(df2['DEWP'])
```

```python
The boxplot representation of the DEWP shows that the data is normally distributed.
```

```python
sns.boxplot(df2['HUMI'])
```

```python
The boxplot representation of the HUMI shows that the data is right skewed..i.e.,From the box and whisker diagram we can also see
that the median is closer to the first quartile than the third quartile.Now I will try to make the right skewed data in normally
disttributed form such that there will be no outliers and whole data will be equally concentrated at a certain point.
```

```python
##Skewness of the data is removed by using the power method and using the cuberoot method.
a =  np.power(df2['HUMI'],(1/3))
sns.kdeplot(a, shade=True, c='b')
```

```python
sns.boxplot(df2['PRES'])
```

```python
sns.kdeplot(df2['PRES'], shade=True, c='b')
```

```python
The above graph for PRES column shows that the data is RIGHT skewwed. Now the skewness can be rempoved by using Power function.
In power function we can square root, cube root or log according to the need.
```

```python
##a1 =  np.power(df2['PRES'],(1/3))
##sns.kdeplot(a1, shade=True, c='b')
a1 = np.log(df2['PRES'])
a1
```

```python
##The graph is now lesser skewed and bit more like normally distributed graph.
a1 =  np.power(df2['PRES'],(a1))
sns.kdeplot(a1, shade=True, c='b')
```
