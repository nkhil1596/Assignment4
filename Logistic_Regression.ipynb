```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\dell\Desktop\NEC\BeijingPM20100101_20151231.csv")
df.head()
```

```python
df1 = df.dropna()
df1.head()
```

```python
dummies_frame = pd.get_dummies(df1)
df2 = df1.reindex(columns = dummies_frame.columns,fill_value =1)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.drop(['year','No','TEMP'],axis = 1)
df3.head()
```

```python
df2_n =df2.select_dtypes(include = [np.number])
df3_n = df3.select_dtypes(include = [np.number])
```

```python
x = np.array(df3_n.loc[:,:])
y = np.array(df2_n.loc[:,'TEMP'])
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets,linear_model
```

```python
X_test,X_train,Y_test,Y_train = train_test_split(x,y,test_size = 0.30,random_state = 7)
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape )
```

```python
from sklearn.linear_model import LogisticRegression
LogisticRegr = LogisticRegression()
LogisticRegr.fit(X_train , Y_train)
```

```python
prediction = LogisticRegr.predict(X_test)
```

```python
a = pd.DataFrame({"actual":list(Y_test), "prediction":list(prediction)})
a
```
