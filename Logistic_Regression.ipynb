```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv")
df.head()
```

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
df1.head()
```

```python
result = [df, df1]
```

```python
result1 = pd.concat(result)
result1
```

```python
df2 = result1.dropna(subset=['Survived'])
df2.head()
```

```python
df2['Has_Cabin']=df2['Cabin'].isnull()
df2.head()
```

```python
import re
df2['Title'] = df2.Name.apply(lambda x:re.search('([A-Z][a-z]+)\.',x).group(1))
df2.head()
```

```python
df2.fillna(df2.median())
```

```python
df2['Categorical_age']= pd.qcut(df2.Age,q=4,labels = False)
df2['Categorical_fare']=pd.qcut(df2.Fare,q=4,labels = False)
df3 = df2.fillna(df2.median())
df3.head()
```

```python
df4 = df3.drop(['Cabin','Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch','Has_Cabin'],axis = 'columns')
df4.head()
```

```python
dummies_frame = pd.get_dummies(df4)
df5= df4.reindex(columns = dummies_frame.columns, fill_value=1)
df5.head()
```

```python
df6 = df5.drop(['Survived'], axis = 1)
df6.head()
```

```python
df6_n= df6.select_dtypes(include = [np.number])
df5_n=df5.select_dtypes(include=[np.number])
```

```python
x= np.array(df6_n.loc[:,:])
y = np.array(df5_n.loc[:,['Survived']])
x.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets, linear_model
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.25,random_state = 0)
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LogisticRegr = LogisticRegression()
LogisticRegr.fit(X_train , Y_train)
```

```python
#Predicts for one observation
LogisticRegr.predict(X_test[0].reshape(1,-1))
```

```python
#Predicts for multiple observation
LogisticRegr.predict(X_test[0:10])
```

```python
#Make predictions for entire dataset
predictions = LogisticRegr.predict(X_test)
```

```python
score =  LogisticRegr.score(X_test,Y_test)
print(score)
```
